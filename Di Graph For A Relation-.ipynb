```python
import networkx as nx
import matplotlib.pyplot as plt

# Function to input a zero-one matrix from the user
def input_matrix():
    n = int(input("Enter the number of nodes: "))  # Size of the matrix (n x n)
    matrix = []
    
    print("Enter the adjacency matrix row by row (0 or 1):")
    for i in range(n):
        row = list(map(int, input(f"Row {i+1}: ").split()))
        matrix.append(row)
    
    return matrix

# Function to create a directed graph from the matrix
def create_directed_graph(matrix):
    G = nx.DiGraph()

    # Adding edges based on the adjacency matrix
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            if matrix[i][j] == 1:
                G.add_edge(i, j)

    return G

# Function to visualize the directed graph
def visualize_graph(G):
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_color='lightblue', arrows=True)
    plt.show()

# Main function
def main():
    matrix = input_matrix()  # Get the user-input matrix
    G = create_directed_graph(matrix)  # Create the directed graph
    visualize_graph(G)  # Visualize the graph

if __name__ == "__main__":
    main()
```
